```python
fig = plt.figure(figsize=(18,10))
# fig.gca().xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m")) #設定x軸主刻度顯示格式
# fig.gca().xaxis.set_major_locator(mdates.DayLocator(interval=500)) #設定x軸主刻度間距

plt.subplot(3,1,1)
x_2330 = df_2330["日期"]
x_2330 = pd.to_datetime(x_2330)
y_2330 = df_2330["收盤"]
plt.plot(x_2330, y_2330, color=(0.2, 0.5, 0.7, 0.4))
plt.legend(['2330'])


plt.subplot(3,1,2)
x_0050 = df_0050["日期"]
x_0050 = pd.to_datetime(x_0050)
y_0050 = df_0050["收盤"]
plt.plot(x_0050, y_0050, color=(0.2, 0.5, 0.2, 0.4))
plt.legend(['0050'])

plt.subplot(3,1,3)
x_00631L = df_00631L["日期"]
x_00631L = pd.to_datetime(x_00631L)
y_00631L = df_00631L["收盤"]
plt.plot(x_00631L, y_00631L, color=(0.7, 0.1, 0.3, 0.4))
plt.legend(['00631L'])


plt.show()
```

```python
fig = plt.figure(figsize=(18,6))

plt.plot(x_2330, y_2330, color=(0.2, 0.5, 0.7, 0.4))
plt.plot(x_0050, y_0050, color=(0.2, 0.5, 0.2, 0.4))
plt.plot(x_00631L, y_00631L, color=(0.7, 0.1, 0.3, 0.4))
plt.legend(['2330','0050','00631L'])


plt.show()
```
